```
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('mtcars.csv')
print(df)
```

```
transmission_counts = df['am'].value_counts()

plt.figure(figsize=(6, 4))
plt.bar(transmission_counts.index, transmission_counts.values, tick_label=['Automatic (0)', 'Manual (1)'])
plt.xlabel('Transmission Type')
plt.ylabel('Frequency')
plt.title('Frequency of Transmission Types')
plt.show()
```

```
plt.figure(figsize=(8, 6))
plt.scatter(df['wt'], df['mpg'])
plt.xlabel('Weight (1000 lbs)')
plt.ylabel('Miles per Gallon (mpg)')
plt.title('Relationship between Car Weight and MPG')
plt.grid(True)
plt.show()
```
